# ENTSO-E Data Fetcher - Hungary Electricity Load Data

## Overview

This notebook fetches historical electricity load data for Hungary from the ENTSO-E (European Network of Transmission System Operators) API. The ENTSO-E platform provides publicly available electricity system data for European countries.

### Data Details

- **Country**: Hungary
- **Time Period**: 2015-2024
- **Temporal Resolution**: 15-minute intervals
- **Data Type**: Actual electricity load (consumption) in MW (Megawatts)
- **Timezone**: UTC (Coordinated Universal Time)

### ENTSO-E API Information

The ENTSO-E Transparency Platform provides:
- Historical load data
- Day-ahead load forecasts
- Generation data
- Transmission data

For this project, we focus on actual load data (document type A65, process type A16) for Hungary's bidding zone.

### Process

1. Configure API connection (API key and bidding zone code)
2. Fetch data month by month (due to API limitations)
3. Parse XML responses and extract time series data
4. Handle timezone conversions
5. Save data to CSV format for further processing


```python
import sys
from pathlib import Path
from datetime import datetime
import importlib

# Proje kÃ¶k dizini
project_root = Path().resolve().parent
sys.path.insert(0, str(project_root / 'src'))

# ModÃ¼lÃ¼ import et
if 'entsoe_data_fetcher' in sys.modules:
    importlib.reload(sys.modules['entsoe_data_fetcher'])

from entsoe_data_fetcher import ENTSOEDataFetcher

print("âœ“ HazÄ±r")
```

```python
fetcher = ENTSOEDataFetcher()
print("âœ… Data fetcher hazÄ±r")
```

```python
# Date range
start_date = datetime(2015, 1, 1)
end_date = datetime(2024, 12, 31)

print(f"ğŸ“… Date range: {start_date.date()} - {end_date.date()}")
```

```python
# Veriyi fetch (ay ay - hÄ±zlÄ±)
df_raw = fetcher.fetch_range(start_date, end_date, show_progress=True)

# Reindex ve interpolation
print("\nğŸ”„ Eksik dataler tespit ediliyor ve dolduruluyor...")
df, df_report = fetcher.reindex_and_fill(df_raw)

if not df_report.empty:
    print(f"\nâš ï¸ {df_report['Num_Points'].sum():,} data noktasÄ± interpolation ile dolduruldu")
    print(f"   {len(df_report)} missing aralÄ±k bulundu")
else:
    print("\nâœ… HiÃ§ missing data yok!")
```

```python
# CSV'ye save
csv_filepath = project_root / "data" / "raw" / "hungary_load_data_2015_2024.csv"
fetcher.save_csv(df, str(csv_filepath))

# Interpolasyon reportnu save
if not df_report.empty:
    report_filepath = project_root / "data" / "raw" / "interpolation_report.csv"
    fetcher.save_csv(df_report, str(report_filepath))
    print(f"âœ… Interpolasyon report saved: {report_filepath}")

print(f"\nğŸ“Š Ã–zet:")
print(f"   KayÄ±t sayÄ±sÄ±: {len(df):,}")
print(f"   Dosya: {csv_filepath}")
print(f"   Boyut: {csv_filepath.stat().st_size / (1024*1024):.2f} MB")
```
